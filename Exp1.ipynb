A. Python code for text handling


```python
text="hello world"
lowercase_text=text.lower()
print(lowercase_text)
```

```python
text="hello world"
words=text.split()
for word in words:
  print (word)
```

```python
text="hello "
reversed_text=text[::-1]
print(reversed_text)
```

```python
text="I like java"
new_text=text.replace("java","python")
print(new_text)
```

```python
text = "Hello World"
letter_count = {}
for letter in text:
    if letter in letter_count:
        letter_count[letter] += 1
    else:
        letter_count[letter] = 1
print(letter_count)

```

```python
from moviepy.editor import VideoFileClip
video_path = next(iter(uploaded))
clip = VideoFileClip(video_path)
print(f"Duration: {clip.duration:.2f} seconds")
print(f"Resolution: {clip.size}")
print(f"FPS: {clip.fps}")
```

```python
short_clip = clip.subclip(0, 5)
short_clip.write_videofile("short_clip.mp4")
```

```python
from moviepy.editor import VideoFileClip
import matplotlib.pyplot as plt
import numpy as np
video_path = next(iter(uploaded))
clip = VideoFileClip(video_path)
frame = clip.get_frame(4.0)
plt.imshow(frame)
plt.title("Frame at 4 Seconds")
plt.axis("off")
plt.show()
```
